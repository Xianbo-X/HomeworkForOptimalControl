```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.patches as patches
from PIL import Image
```

```python
def getcontrolpoints(letter:str):
    map={
        "A": [np.array([[6,16,31,39,54,64],[77,51,6,6,51,77]]),np.array([[1,2,3,4,5,5],[2,3,4,5,2,6]])]
    }
    return map[letter]

def Imshow(fig,image,cmap,title=None):
    ax=fig.add_subplot(1,1,1)
    ax.imshow(image,cmap)
    ax.axis("off")
    if title!=None:
        ax.set_title(title)
```

```python
ctrpoints,links=getcontrolpoints("A")
```

```python
im=Image.open("AllLetters.jpg")
im2=im.convert("L")
Imshow(plt.figure(),im2,"gray")
bn=np.array(np.array(im2)<100).astype(np.int64)
Imshow(plt.figure(),bn,"gray")
bnp=np.zeros(bn.shape)
Imshow(plt.figure(),bnp,"gray","Empty background")
```
